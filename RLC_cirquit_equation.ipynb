# Анализ размерностей

## Приведение уравнения параллельной RLC цепи к чисто математической задаче

Ток в каждом из компонентов параллельной RLC цепи:

$$I_C=\frac{\texttt{d}q}{\texttt{d}t} = C\frac{\texttt{d}U}{\texttt{d}t},\quad
I_R=\frac{U}{R}, \quad
I_L=\frac{1}{L}\int U\texttt{d}t$$

Закон Кирхгофа (сумма токов в узле):

$$I_\text{src} = I_C+I_R+I_L =
C\frac{\texttt{d}U}{\texttt{d}t}+\frac{U}{R}+\frac{1}{L}\int U\texttt{d}t$$

Дифференциальное уравнение для напряжения в параллельной RLC цепи:
$$\frac{\texttt{d}^2U}{\texttt{d}t^2}+\frac{1}{RC}\frac{\texttt{d}U}{\texttt{d}t}+\frac{1}{LC}U=\frac{1}{C}\frac{\texttt{d}I_\text{src}}{\texttt{d}t}$$
Можно и для заряда, т.к. он пропорционален напряжению:
$$\frac{\texttt{d}^2q}{\texttt{d}t^2}+\frac{1}{RC}\frac{\texttt{d}q}{\texttt{d}t}+\frac{1}{LC}q=\frac{\texttt{d}I_\text{src}}{\texttt{d}t}$$

Введём обозначения для параметров:

$$\omega_0^2=\frac{1}{LC},\quad
\frac{1}{RC}=\frac{1}{R}\sqrt{\frac{L}{C}}\frac{1}{\sqrt{LC}}=\frac{\omega_0}{Q},\quad
Q=R\sqrt{\frac{C}{L}}$$

$$\frac{\texttt{d}^2U}{\texttt{d}t^2}+\frac{\omega_0}{Q}\frac{\texttt{d}U}{\texttt{d}t}+\omega_0^2U=\frac{1}{C}\frac{\texttt{d}I_\text{src}}{\texttt{d}t}$$

Разделим на $\omega_0^2$, чтобы продолжить процесс обезразмеривания уравнения дальше, и свести задачу к чисто математической. Безразмерное время: $\hat{t}=\omega_0t$. Получившийся коэффициент в правой части уравнения можно представить в более удобной форме:

$$\frac{1}{C\omega_0}=\frac{1}{R}\sqrt{\frac{L}{C}}\cdot R = \frac{R}{Q}$$

В данном случае производные по безразмерному времени, поэтому размерность каждого слагаемого и правой части — вольты.

$$\frac{\texttt{d}^2U}{\texttt{d}\hat{t}^2}+\frac{1}{Q}\frac{\texttt{d}U}{\texttt{d}\hat{t}}+U=\frac{R}{Q}\frac{\texttt{d}I_\text{src}}{\texttt{d}\hat{t}}$$

Обозначим размерные величины через размерный коэффициент и безразмерные функции $x(\hat{t})$, $y(\hat{t})$:

$$U=U_0y(\hat{t}),\quad I_\text{src}=I_0x(\hat{t})$$

Размерные коэффициенты можно сократить, если принять $U_0=RI_0$. Это можно сделать, так как количество введённых параметров явно избыточно, и их придётся либо связать друг с другом в процессе решения, либо наложить ограничения заранее, но потом использовать их при решении. Выбор такого соотношения также удобен благодаря свойствам последовательной RLC-цепи: $R$ — это одновременно и эквивалентное сопротивление цепи в резонансе.
$$\frac{\texttt{d}^2y}{\texttt{d}\hat{t}^2}+\frac{1}{Q}\frac{\texttt{d}y}{\texttt{d}\hat{t}}+y=\frac{1}{Q}\frac{\texttt{d}x}{\texttt{d}\hat{t}}$$
На этом с физикой полностью покончено, задача сведена к чисто математической.

## Приведение уравнения последовательной RLC цепи к чисто математической задаче

Напряжение на каждом из компонентов последовательной RLC цепи:
$$q = CU_C=\int Idt,\quad U_R = RI, \quad U_C = \frac{1}{C}\int Idt, \quad U_L = L\frac{\texttt{d}I}{\texttt{d}t}$$
Закон Кирхгофа:
$$U_\text{src} = U_R+U_L+U_C = RI + L\frac{\texttt{d}I}{\texttt{d}t} + \frac{1}{C}\int Idt$$
Дифференциальное уравнение для тока в последовательной RLC цепи:
$$\frac{\texttt{d}^2I}{\texttt{d}t^2}+\frac{R}{L}\frac{\texttt{d}I}{\texttt{d}t}+\frac{1}{LC}I=\frac{1}{L}\frac{\texttt{d}U_\text{src}}{\texttt{d}t}$$
Введём обозначения для параметров:
$$\omega_0^2=\frac{1}{LC},\quad \frac{R}{L}=\frac{1}{\sqrt{LC}}R\sqrt{\frac{C}{L}}=\frac{\omega_0}{Q},\quad Q=\frac{1}{R}\sqrt{\frac{L}{C}}$$
$$\frac{\texttt{d}^2I}{\texttt{d}t^2}+\frac{\omega_0}{Q}\frac{\texttt{d}I}{\texttt{d}t}+\omega_0^2I=\frac{1}{L}\frac{\texttt{d}U_\text{src}}{\texttt{d}t}$$
Разделим на $\omega_0^2$, чтобы продолжить процесс обезразмеривания уравнения дальше, и свести задачу к чисто математической. Безразмерное время: $\hat{t}=\omega_0t$. Получившийся коэффициент в правой части уравнения можно представить в более удобной форме:
$$\frac{1}{L\omega_0}=\sqrt{\frac{C}{L}}=\frac{1}{QR}$$
В данном случае производные по безразмерному времени, поэтому размерность каждого слагаемого и правой части — амперы.
$$\frac{\texttt{d}^2I}{\texttt{d}\hat{t}^2}+\frac{1}{Q}\frac{\texttt{d}I}{\texttt{d}\hat{t}}+I=\frac{1}{QR}\frac{\texttt{d}U_\text{src}}{\texttt{d}\hat{t}}$$
Обозначим размерные величины через размерный коэффициент и безразмерные функции $x(\hat{t})$, $y(\hat{t})$:
$$I=I_0y(\hat{t}),\quad U_\text{src}=U_0x(\hat{t})$$
Размерные коэффициенты можно сократить, если принять $I_0=U_0/R$. Это можно сделать, так как количество введённых параметров явно избыточно, и их придётся либо связать друг с другом в процессе решения, либо наложить ограничения заранее, но потом использовать их при решении. Выбор такого соотношения также удобен благодаря свойствам последовательной RLC-цепи: $R$ — это одновременно и эквивалентное сопротивление цепи в резонансе.
$$\frac{\texttt{d}^2y}{\texttt{d}\hat{t}^2}+\frac{1}{Q}\frac{\texttt{d}y}{\texttt{d}\hat{t}}+y=\frac{1}{Q}\frac{\texttt{d}x}{\texttt{d}\hat{t}}$$

## Цепочка Вина

![Wein's circuit](weins-circuit.png)

Напряжение $U_\text{in}$ — сумма напряжений на последовательных элементах, $U_\text{out}$ — напряжение на последнем участке из параллельно соединённых элементов:  

$$U_\text{in} = RI_1+\frac{1}{C}\int I_1 dt + U_\text{out}$$

Продифференцировав это уравнение и подставив в него ток 

$$I_1=I_2+I_3=C\frac{\texttt{d}U_\text{out}}{\texttt{d}t}+\frac{1}{R}U_\text{out}$$

получаем дифференциальное уравнение второго порядка:

$$RC\frac{\texttt{d}^2U_\text{out}}{\texttt{d}t^2}+3\frac{\texttt{d}U_\text{out}}{\texttt{d}t}+\frac{1}{RC}U_\text{out}=\frac{\texttt{d}U_\text{in}}{\texttt{d}{t}}$$

Разделим на $RC$ и ведём такие обозначения для параметров:

$$Q=\frac{1}{3},\quad \omega_0=\frac{1}{RC}$$

чтобы получить уравнение аналогичное тому, что было для колебательных контуров

$$\frac{\texttt{d}^2U_\text{out}}{\texttt{d}t^2}+\frac{\omega_0}{Q}\frac{\texttt{d}U_\text{out}}{\texttt{d}t}+\omega_0^2U_\text{out}=\omega_0\frac{\texttt{d}U_\text{in}}{\texttt{d}{t}}$$

разделим на $\omega_0$ и введём безразмерное время $\hat{t}=\omega_0 t$

$$\frac{\texttt{d}^2U_\text{out}}{\texttt{d}\hat{t}^2}+\frac{1}{Q}\frac{\texttt{d}U_\text{out}}{\texttt{d}\hat{t}}+U_\text{out}=\frac{\texttt{d}U_\text{in}}{\texttt{d}\hat{t}}$$

Выразим напряжения через безразмерные функции и размерные коэффициенты:

$$U_\text{in}=U_{\text{out},0}x(\hat{t}),\quad U_\text{out}=U_{0} y(\hat{t})$$

Если амплитуды связаны соотношением $U_{\text{out},0}=QU_{0}$, где добротность $Q=1/3$ — безразмерная величина, то снова получаем аналогичное уравнение (правую часть умножили и разделили на $Q$):

$$\frac{\texttt{d}^2y}{\texttt{d}\hat{t}^2}+\frac{1}{Q}\frac{\texttt{d}y}{\texttt{d}\hat{t}}+y=\frac{1}{Q}\frac{\texttt{d}x}{\texttt{d}\hat{t}}$$

## Заключение

Обе задачи свелись к одному математическому уравнению. Разница только в размерных коэффициентах. В этом смысле задачи подобны.

Параллельное соединение:

$$ Q=R\sqrt{\frac{C}{L}},\quad \omega_0=\frac{1}{\sqrt{LC}},\quad U(t) = RI_0 y(\omega_0 t)$$

Последовательная цепь:

$$Q=\frac{1}{R}\sqrt{\frac{L}{C}},\quad \omega_0=\frac{1}{\sqrt{LC}},\quad I(t) = \frac{U_0}{R} y(\omega_0 t)$$

Частота свободных колебаний в отсутствии затухания $\omega_0$, а также решение характеристического уравнения одинаковы.

Цепочка Вина:

$$Q=\frac{1}{3},\quad \omega_0=\frac{1}{RC},\quad U_\text{out}(t)=QU_0y(\omega_0t)$$

или более привычной для RC-цепей форме:

$$U_\text{out}(t)=\frac{1}{3} U_0 y(t/\tau),\quad \tau=RC$$
